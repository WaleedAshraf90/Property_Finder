```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time as t
import csv
import random
import threading as th
```

```python
driver= webdriver.Chrome() 
```

```python
Type_lst=[]
Title_lst=[]
Location_lst=[]
Bedrooms_lst=[]
Bathrooms_lst=[]
Spaces_lst=[]
Price_lst=[]

for page in range(2,3):
    t.sleep(2)
    driver.get(f'https://www.propertyfinder.eg/en/search?c=1&ob=mr&page={page}')
    t.sleep(1)
    # Scrap_type: 
    containers= driver.find_elements_by_css_selector("div[class='card__body']") 
    for container in containers:
        try:
            Type= container.find_element_by_css_selector("p[class='card-intro__type']").text 
            Type_lst.append(Type) 
        except:
            Type_lst.append("")

            
    # Scrap_Title: 
    containers= driver.find_elements_by_css_selector("div[class='card__body']") 
    for container in containers:
        try:
            Title= container.find_element_by_css_selector("h2[class='card-title__title']").text 
            Title_lst.append(Title)
        except:
            Title_lst.append("")


    # Scrap_Location: 
    containers= driver.find_elements_by_css_selector("div[class='card__body']") 
    for container in containers:
        try:
            Location= container.find_element_by_css_selector("span[class='card__location-text']").text 
            Location_lst.append(Location)
        except:
            Location_lst.append("")


    # scrap_Bedrooms:
    containers= driver.find_elements_by_css_selector("div[class='card__body']")        
    for container in containers:
        try:
            Bedrooms= container.find_element_by_css_selector("div[class='card-specifications__amenities']").find_element_by_css_selector("p:nth-child(1)").text
            Bedrooms_lst.append(Bedrooms)
        except:
            Bedrooms_lst.append("")


    # scrap_bathrooms:
    containers= driver.find_elements_by_css_selector("div[class='card__body']")        
    for container in containers:
        try:
            Bathrooms= container.find_element_by_css_selector("div[class='card-specifications__amenities']").find_element_by_css_selector("p:nth-child(2)").text
            Bathrooms_lst.append(Bathrooms)
        except:
            Bathrooms_lst.append(" ")


    # scrap_Space:
    containers= driver.find_elements_by_css_selector("div[class='card__body']")        
    for container in containers:
        try:
            Spaces= container.find_element_by_css_selector("div[class='card-specifications__amenities']").find_element_by_css_selector("p:nth-child(3)").text
            Spaces_lst.append(Spaces)
        except:
            Spaces_lst.append(" ")     


    # Scrap_price: 
    containers= driver.find_elements_by_css_selector("div[class='card__body']") 
    for container in containers:
        try:
            Price= container.find_element_by_css_selector("p[class='card-intro__price']").text 
            Price_lst.append(Price) 
        except:
            Price_lst.append(" ")


####################################################################
from itertools import zip_longest
from csv import writer

file_list= [Type_lst, Title_lst, Location_lst, Bedrooms_lst, Bathrooms_lst, Spaces_lst, Price_lst]
exported= zip_longest(*file_list)

with open(f'propertyfindertest.csv', 'w', encoding="utf-8") as csv_file:
    wr= writer(csv_file)
    wr.writerow(['Type','Title','Location','Bedrooms','Bathrooms','Space','Price'])
    wr.writerows(exported) 
```
